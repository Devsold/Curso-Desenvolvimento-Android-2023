```python
import numpy as np
import pandas as pd

```

```python
#@title
df=pd.read_csv('agricultural_raw_material.csv')
```

```python
df.info
df.isnull().sum()
```

```python
df = df.replace('%', '', regex=True)
df = df.replace(',', '', regex=True)
df = df.replace('-', '', regex=True)
df = df.replace('MAY90', np.nan)
df = df.dropna()
df.isnull().sum()
lst = ['Coarse wool Price', 'Coarse wool price % change',"Copra Price", 'Copra price % change', 'Cotton price % Chance','Fine wool Price','Fine wool Price % Chance','Hard log price % Chance']
df[lst] = df[lst].astype('float')
df.dtypes
```

```python
df.head()
```

```python
df.Month = pd.to_datetime(df.Month.str.upper(), format='%b%y', yearfirst=False)
df = df.set_index('Mounth')
```

```python
df.head()
```

```python
from ast import increment_lineno
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (0, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python

raw_data = ['Coarse wool Price', 'Copra Price', 'Cotton Price', 'Fine wool Price', 'Hard log Price', 'Hard sawnwood Price',
            'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 'Soft sawnwood Price', 'Wood pulp Price']
corrmat = df[raw_data].corr()
fig = plt.figure(figsize = (12, 9)) # Decrease the indentation of this line by one level
mask = np.triu(np.ones_like(corrmat, dtype=bool))
sns.heatmap(corrmat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```
